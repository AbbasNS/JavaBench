# Completion(Class-wise)

```python
from visual_table import PA, models, get_lazy_data, concat_df, style_df
def get_df_completion(n = 5, k = 1):
    return concat_df([
        get_lazy_data(PA, models, label="incremental-selective", n = n, k = k),
        get_lazy_data(PA, models, label="incremental_random-selective", n = n, k = k),
        get_lazy_data(PA, models, label="incremental_rev-selective", n = n, k = k),
    ])
df_completion = get_df_completion()
style_df(df_completion)
```

# Merged

```python
from visual_table import join_all
from IPython.display import display

def get_merge(n = 5, k = 1):
    return join_all(
        get_df_completion(n, k),
        get_df_compilation_class_wise(n, k),
        get_df_compilation_test_wise(n, k),
        get_df_pass_class_wise(n, k),
        get_df_pass_test_wise(n, k),
        orders=['Model', 'Method'],
    )

display(get_merge(5, 1))
display(get_merge(5, 3))
display(get_merge(5, 5))
```

```python
import json

def process(df):
    df["Pass_at_1"] = df[['PA19', 'PA20', 'PA21', 'PA22']].mean(axis=1)
    df = df.drop(columns=['PA19', 'PA20', 'PA21', 'PA22'])
    return json.loads(df.to_json(orient='records'))

with open('data/data_incr-order.json', 'w') as file:
    p1 = process(get_df_completion(5, 1))
    p3 = process(get_df_compilation_class_wise(5, 1))
    p4 = process(get_df_compilation_test_wise(5, 1))
    p5 = process(get_df_pass_class_wise(5, 1))
    p6 = process(get_df_pass_test_wise(5, 1))
    json_str = json.dumps({
        'completion': p1,
        'compilation_class_wise': p3,
        'compilation_test_wise': p4,
        'pass_class_wise': p5,
        'pass_test_wise': p6,
    }, indent=2)
    file.write(json_str)
```
